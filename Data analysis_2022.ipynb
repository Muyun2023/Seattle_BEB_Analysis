```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "2022_Building_Energy_Benchmarking_20240322.csv"
data = pd.read_csv(file_path)

data.head(), data.info()
```

```python
columns_to_check = [
    'ENERGYSTARScore', 'SiteEUIWN(kBtu/sf)', 'SiteEUI(kBtu/sf)',
    'SiteEnergyUse(kBtu)', 'SiteEnergyUseWN(kBtu)', 'SourceEUIWN(kBtu/sf)',
    'SourceEUI(kBtu/sf)']

# IQR
outliers_summary = {}

for column in columns_to_check:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_count = data.loc[(data[column] < lower_bound) | (data[column] > upper_bound), column].count()
    outliers_summary[column] = outliers_count

outliers_df = pd.DataFrame.from_dict(outliers_summary, orient='index', columns=['Outliers Count'])
outliers_df
```

```python
# bloxplot
# Set the style of seaborn
sns.set_style("whitegrid")

# Creating boxplots for 'SiteEnergyUse(kBtu)' and 'ENERGYSTARScore'
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Site Energy Use (kBtu)
sns.boxplot(data=data, y='SiteEnergyUse(kBtu)', ax=ax[0])
ax[0].set_title('Site Energy Use (kBtu)')
ax[0].set_ylabel('Energy Use (kBtu)')

# ENERGYSTARScore
sns.boxplot(data=data, y='ENERGYSTARScore', ax=ax[1])
ax[1].set_title('ENERGYSTARScore')
ax[1].set_ylabel('Score')

plt.tight_layout()
plt.show()
```

```python
# compliance status distribution 'EPAPropertyType' and 'ComplianceStatus'
compliance_status_grouped = data.groupby(['EPAPropertyType', 'ComplianceStatus']).size().unstack(fill_value=0)
compliance_status_grouped
```

```python
compliance_status_percentage = compliance_status_grouped.apply(lambda x: (x / x.sum()) * 100, axis=1)

compliance_status_percentage_sorted = compliance_status_percentage.sort_values(by='Not Compliant', ascending=False)
compliance_status_percentage_sorted
```

```python
# Calculate the frequency of each category in 'ComplianceIssue' column
compliance_issue_freq = data['ComplianceIssue'].value_counts()

# Display the frequency distribution
compliance_issue_freq
```

```python
# Anomaly Detection for 'SiteEnergyUse(kBtu)' using Z-score to identify outliers
from scipy.stats import zscore

# Compute Z-scores for 'SiteEnergyUse(kBtu)'
data['SiteEnergyUse_zscore'] = zscore(data['SiteEnergyUse(kBtu)'].dropna())

# Identify outliers as those with a Z-score greater than 3 or less than -3 (typical threshold for outliers)
outliers_energy_use = data.loc[data['SiteEnergyUse_zscore'].abs() > 3]

# Consistency Check for Gross Floor Area
# Checking if the sum of 'PropertyGFABuilding(s)' and 'PropertyGFAParking' equals 'PropertyGFATotal'
data['GFA_Check'] = (data['PropertyGFABuilding(s)'] + data['PropertyGFAParking']) == data['PropertyGFATotal']

# Summary of findings
outliers_summary = {
    'Outliers in SiteEnergyUse(kBtu)': outliers_energy_use.shape[0],
    'Inconsistencies in GFA Reporting': (~data['GFA_Check']).sum()
}

print(outliers_summary)

# Display the rows corresponding to the 30 outliers identified in 'SiteEnergyUse(kBtu)'
outliers_energy_use[['OSEBuildingID', 'BuildingName', 'SiteEnergyUse(kBtu)']]

#SWEDISH MEDICAL CENTER CHERRY HILL - CAMPUS (OSEBuildingID: 268) is listed as "Not Compliant."
```

```python
file_path = "2021_Building_Energy_Benchmarking_20240322.csv"
file_path2 = "2022_Building_Energy_Benchmarking_20240322.csv"
data_2021 = pd.read_csv(file_path)
data_2022 = pd.read_csv(file_path2)
```

```python
## Compare the missing rates between 2021 and 2022 datasets, and identify that the data collecion lagging exists, though the 2021 data is finalized by now,
## the 2022 data is still in progress and have many missing values.

# Define the columns for which we want to find the missing value rate
columns_of_interest = [
    'ENERGYSTARScore', 'SiteEUIWN(kBtu/sf)', 'SiteEUI(kBtu/sf)', 'SiteEnergyUse(kBtu)',
    'SiteEnergyUseWN(kBtu)', 'SourceEUIWN(kBtu/sf)', 'SourceEUI(kBtu/sf)',
    'TotalGHGEmissions', 'GHGEmissionsIntensity'
]

# Calculate the missing value rate for each column in both 2021 and 2022 datasets
missing_rate_2021 = data_2021[columns_of_interest].isnull().mean() * 100
missing_rate_2022 = data_2022[columns_of_interest].isnull().mean() * 100

# Create a DataFrame to compare the missing value rates side by side
missing_rate_comparison = pd.DataFrame({
    '2021 Missing Rate (%)': missing_rate_2021,
    '2022 Missing Rate (%)': missing_rate_2022
})

missing_rate_comparison
```

```python
filtered_data = data[data['TotalGHGEmissions'].isnull()]

# Grouping the data by 'Neighborhood' and 'ComplianceIssue' to understand the frequency of compliance issues across neighborhoods
neighborhood_compliance_issue_freq = filtered_data.groupby(['Neighborhood', 'ComplianceIssue']).size().unstack(fill_value=0)

# Display the grouped data
neighborhood_compliance_issue_freq
```

```python
# Calculating the percentage of missing and non-missing values for 'GHGEmissionsIntensity'
ghg_missing_percentage = data['GHGEmissionsIntensity'].isnull().mean() * 100
ghg_present_percentage = 100 - ghg_missing_percentage

# Preparing data for the pie chart
ghg_data = [ghg_present_percentage, ghg_missing_percentage]
ghg_labels = ['Data Present', 'Data Missing']

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(ghg_data, labels=ghg_labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Percentage of Missing vs. Present GHG Emissions Intensity Data in 2022')

plt.show()
```

```python

import matplotlib.pyplot as plt

# Sort the neighborhoods by the total count of missing data
neighborhoods_sorted = neighborhood_compliance_issue_freq.sum(axis=1).sort_values(ascending=False)

# Plot the bar chart and add % labels to the bars
plt.figure(figsize=(12, 6))
ax = neighborhoods_sorted.plot(kind='bar')
plt.title('Frequency of Neighborhoods in Missing Data')
plt.xlabel('Neighborhood')
plt.ylabel('Frequency')
plt.xticks(rotation=90)

# Add the numbers to the bars
for i, v in enumerate(neighborhoods_sorted):
    ax.text(i, v, str(v), ha='center', va='bottom')

plt.show()

```
